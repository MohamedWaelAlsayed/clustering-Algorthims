## Importing Libraries

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

## Reading Dataset

```python
test = pd.read_csv("Iris.csv")
data = test.copy()
```

## preview Data

```python
data.head()
```

```python
data.info()
```

## remove id attribute

```python
data = data.drop("Id",axis=1)
data.info()
```

```python
data.head()
```

## remove labels

```python
x = data.iloc[:,:4]
y = data.iloc[:,4]
```

```python
x.info()
```

## Drawing dendogram

```python
dendrogram =sch.dendrogram(sch.linkage(x,"single"))
plt.title('dendrogram')
plt.xlabel('x values')
plt.ylabel('distance')
plt.show()
```

## Aglomerative clustering (hierarchical)

```python
cluster_H = AgglomerativeClustering(n_clusters=2,affinity="euclidean",linkage="single")
fitted_H = cluster_H.fit_predict(x)
fitted_H
```

## Visualization

```python
plt.scatter(x.iloc[:,0],x.iloc[:,1],c=cluster_H.labels_)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Aglomerative clustering (hierarchical)")
plt.show()
```

```python
plt.scatter(x.iloc[:,0],x.iloc[:,2],c=cluster_H.labels_)
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.title("Aglomerative clustering (hierarchical)")
plt.show()
```

```python
plt.scatter(x.iloc[:,2],x.iloc[:,3],c=cluster_H.labels_)
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Aglomerative clustering (hierarchical)")
plt.show()
```
